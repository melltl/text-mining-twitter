pre-processamento de dados

```python
def RemoveStopWords(instancia):
  #base de dados da nltk em pt
  stopwords=set(nltk.corpus.stopwords.words('portuguese'))
  #para frase de exemplo separa por espaços e se ela n for stopwords retorna
  palavras=[i for i in instancia.split() if not i in stopwords]
  return(" ".join(palavras))
```

```python
def Stemming(instancia):
  stemmer=nltk.stem.RSLPStemmer()
  palavras=[]
  for w in instancia.split():
    palavras.append9stemmer.stem(w)
  return(" ".join(palavras))
```

```python
def Limpeza_dados(instancia):
  #remove links,pontos,virgulas
  instancia= re.sub(r"http\S+","",instancia).lower().replace(';','').replace('-','').replace(':','').replace(')','')
```

aplicando as 3 funcoes

```python
def Preprocessing(instancia):
  stemmer=nltk.stem.RSLPStemmer()
  instancia=re.sub(r"http\S+","",instancia).lower().replace(';','').replace('-','').replace(':','').replace(')','')
  stopwords=set(nltk.corpus.stopwords.words('portuguese'))
  palavras=[stemmer.stem(i) for i in instancia.split() if not i in stopwords]
  return(" ".join(palavras))

tweets=[Preprocessing(i) for i in tweets]
tweets[:50]
```

criando modelo

funcao de tags de negação

tem como objetivo dar mais peso para o modelo identificar uma inversãoo de sentimento da frase


```python
def marque_negacao(texto):
  negacoes=['não','not']
  negacao_detectada=False
  resultado=[]
  palavras=texto.split()
  for p in palavras:
    p=p.lower()
    if negacao_detectada==True:
      p=p+'_NEG'
    if p in negacoes:
      negacao_detectada==True
    resultado.append(p)
  return(" ".join(resultado))
```

criando modelos com pipeline(reduzir codigos  automatizr fluxos)

```python
from sklearn.pipeline import Pipeline
```

- count vectorize:
    - transforma os textos em uma matriz
    - cada linha representa um documento
    - e cada coluna uma palavra unica
    - o valor de cada celula indica a frequencia que as palavras aparecem

- MultinomialNB:
 -it learns the probability distribution of the words in the documents and predicts the most likely class label for new documents

```python
pipeline_simples=Pipeline([
    ('counts',CountVectorizer()),
    ('classifier',MultinomialNB())
])
```

```python
#pipeline de tag de negacao
pipeline_negacoes=Pipeline([
    ('counts',CountVectorizer(tokenizer=lambda text: marque_negacao(text))),
    ('classifier',MultinomialNB())
])
```

```python
pipeline_simples.fit(tweets,classes)
```

```python
pipeline_simples.steps
```

```python
pipeline_negacoes.fit(tweets,classes)
```

```python
pipeline_negacoes.steps
```

validando modelos com validação cruzada


fazendo o cross calidation model

```python
resultados=cross_val_predict(pipeline_simples,tweets,classes,cv=10)
```

```python
metrics.accuracy_score(classes,resultados)
```

```python
sentimento=['Positivo','Negativo','Neutro']
print(metrics.classification_report(classes, resultados, target_names=sentimento))
```

```python
print(pd.crosstab(classes,resultados,rownames=['Real'],colnames=['Predito'],margins=True))
```
